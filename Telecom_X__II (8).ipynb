## **Medidas_de_Dispersion**
Las medidas de dispersiÃ³n en estadÃ­stica describen la variabilidad o extensiÃ³n de los datos alrededor de un valor central, como la media aritmÃ©tica. Indican cuÃ¡n dispersos o agrupados estÃ¡n los datos. Algunas medidas comunes de dispersiÃ³n son el rango, la desviaciÃ³n estÃ¡ndar, la varianza, el rango intercuartil, la desviaciÃ³n media absoluta, y el coeficiente de variaciÃ³n.

A diferencia de las medidas de tendencia central (media, mediana, moda), que solo indican el centro de los datos, las medidas de dispersiÃ³n revelan cÃ³mo de dispersos o agrupados estÃ¡n los datos con respecto a ese centro.

----//Varianza//---------

La varianza es una medida estadÃ­stica que indica la dispersiÃ³n de los datos en un conjunto. Se calcula como el promedio de los cuadrados de las desviaciones de cada valor respecto a la media. En esencia, muestra cuÃ¡n dispersos estÃ¡n los datos alrededor de su valor promedio. Una varianza alta implica mayor dispersiÃ³n, mientras que una varianza baja indica que los datos estÃ¡n mÃ¡s agrupados cerca de la media.

```python
Varianza_X=baseX[['charges.monthly',	'charges.total',	'Tiempo_de_Contrato',	'cuentas_diarias',	'total_contracto']].var()
Varianza_X
```

----//DesviaciÃ³n_Estandar//---------
La desviaciÃ³n estÃ¡ndar es una medida estadÃ­stica que indica quÃ© tan dispersos estÃ¡n los datos con respecto a su media o promedio. En otras palabras, nos dice si los valores individuales de un conjunto de datos tienden a estar cerca de la media o si estÃ¡n muy dispersos. Una desviaciÃ³n estÃ¡ndar baja indica que los datos estÃ¡n muy agrupados alrededor de la media, mientras que una desviaciÃ³n estÃ¡ndar alta indica una mayor dispersiÃ³n.

##**AnÃ¡lisis de CorrelaciÃ³n**

correlacion_X

La correlaciÃ³n estadÃ­stica es una medida que expresa el grado de relaciÃ³n lineal entre dos variables numÃ©ricas. Indica si al cambiar una variable, la otra tiende a cambiar en la misma direcciÃ³n (correlaciÃ³n positiva) o en direcciÃ³n opuesta (correlaciÃ³n negativa). No implica causalidad, solo indica la fuerza y direcciÃ³n de la relaciÃ³n.
Tipos de correlaciÃ³n:
CorrelaciÃ³n positiva: Las variables aumentan o disminuyen juntas.
CorrelaciÃ³n negativa: Una variable aumenta mientras la otra disminuye.
CorrelaciÃ³n nula: No hay relaciÃ³n aparente entre las variables.
MediciÃ³n de la correlaciÃ³n:
Coeficiente de correlaciÃ³n de Pearson: Mide la relaciÃ³n lineal entre dos variables continuas.
Coeficiente de correlaciÃ³n de Spearman: Mide la relaciÃ³n entre las posiciones relativas de los datos, Ãºtil para datos ordinales o no normales.

Correlacion_X=
InterpretaciÃ³n:
El coeficiente de correlaciÃ³n varÃ­a entre -1 y 1.
Un valor cercano a 1 o -1 indica una correlaciÃ³n fuerte.
Un valor cercano a 0 indica una correlaciÃ³n dÃ©bil.
La magnitud del efecto (fuerza de la correlaciÃ³n) puede interpretarse segÃºn criterios como el de Cohen (0.1-0.3 pequeÃ±o, 0.3-0.5 medio, >0.5 grande).
Importancia:
La correlaciÃ³n es una herramienta fundamental en el anÃ¡lisis de datos, ya que permite:
Entender las relaciones entre variables.
Identificar patrones y tendencias.
Predecir el comportamiento de una variable a partir de otra.
Tomar decisiones informadas basadas en la relaciÃ³n entre variables.
