## **Medidas_de_Dispersion**
Las medidas de dispersión en estadística describen la variabilidad o extensión de los datos alrededor de un valor central, como la media aritmética. Indican cuán dispersos o agrupados están los datos. Algunas medidas comunes de dispersión son el rango, la desviación estándar, la varianza, el rango intercuartil, la desviación media absoluta, y el coeficiente de variación.

A diferencia de las medidas de tendencia central (media, mediana, moda), que solo indican el centro de los datos, las medidas de dispersión revelan cómo de dispersos o agrupados están los datos con respecto a ese centro.

----//Varianza//---------

La varianza es una medida estadística que indica la dispersión de los datos en un conjunto. Se calcula como el promedio de los cuadrados de las desviaciones de cada valor respecto a la media. En esencia, muestra cuán dispersos están los datos alrededor de su valor promedio. Una varianza alta implica mayor dispersión, mientras que una varianza baja indica que los datos están más agrupados cerca de la media.

```python
Varianza_X=baseX[['charges.monthly',	'charges.total',	'Tiempo_de_Contrato',	'cuentas_diarias',	'total_contracto']].var()
Varianza_X
```

----//Desviación_Estandar//---------
La desviación estándar es una medida estadística que indica qué tan dispersos están los datos con respecto a su media o promedio. En otras palabras, nos dice si los valores individuales de un conjunto de datos tienden a estar cerca de la media o si están muy dispersos. Una desviación estándar baja indica que los datos están muy agrupados alrededor de la media, mientras que una desviación estándar alta indica una mayor dispersión.

```python
Desviacion_Estandar_X=baseX[['charges.monthly',	'charges.total',	'Tiempo_de_Contrato',	'cuentas_diarias',	'total_contracto']].std()
Desviacion_Estandar_X
```

##**Análisis de Correlación**

correlacion_X

La correlación estadística es una medida que expresa el grado de relación lineal entre dos variables numéricas. Indica si al cambiar una variable, la otra tiende a cambiar en la misma dirección (correlación positiva) o en dirección opuesta (correlación negativa). No implica causalidad, solo indica la fuerza y dirección de la relación.
Tipos de correlación:
Correlación positiva: Las variables aumentan o disminuyen juntas.
Correlación negativa: Una variable aumenta mientras la otra disminuye.
Correlación nula: No hay relación aparente entre las variables.
Medición de la correlación:
Coeficiente de correlación de Pearson: Mide la relación lineal entre dos variables continuas.
Coeficiente de correlación de Spearman: Mide la relación entre las posiciones relativas de los datos, útil para datos ordinales o no normales.

Correlacion_X=
Interpretación:
El coeficiente de correlación varía entre -1 y 1.
Un valor cercano a 1 o -1 indica una correlación fuerte.
Un valor cercano a 0 indica una correlación débil.
La magnitud del efecto (fuerza de la correlación) puede interpretarse según criterios como el de Cohen (0.1-0.3 pequeño, 0.3-0.5 medio, >0.5 grande).
Importancia:
La correlación es una herramienta fundamental en el análisis de datos, ya que permite:
Entender las relaciones entre variables.
Identificar patrones y tendencias.
Predecir el comportamiento de una variable a partir de otra.
Tomar decisiones informadas basadas en la relación entre variables.
